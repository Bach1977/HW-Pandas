```python
# Import the package 
import pandas as pd
import numpy as np
```

```python
# Import the data file 
School = pd.read_csv('../PyCitySchools/raw_data/schools_complete.csv')
Student = pd.read_csv('../PyCitySchools/raw_data/students_complete.csv')
```

```python

# Some Preliminary analysis to do Q/A of the data
School.info()
Student.info()
```

```python
Total_students = len(Student['Student ID'])
Total_budget = School['budget'].sum()
Av_reading = Student['reading_score'].mean()
Av_math = Student['math_score'].mean()
Passing_math_percent= (len(Student.loc[Student['math_score']>65])/Total_students)*100
Passing_reading_percent= (len(Student.loc[Student['reading_score']>65])/Total_students)*100
Overall_passing = (Passing_math_percent + Passing_reading_percent)/2             
```

```python
# create a summary table
District_Summary = pd.DataFrame({
        
                'Total Students':[Total_students],
                'Total Budget':[Total_budget],
                'Average Reading':[Av_reading],
                'Average Math':[Av_math],
                '% Passing Math':[Passing_math_percent],
                '% Passing reading':[Passing_reading_percent],
                '% Overal Passing ':[Overall_passing]})

District_Summary = District_Summary[['Total Students',
                'Total Budget',
                'Average Reading',
                'Average Math',
                '% Passing Math',
                '% Passing reading',
                '% Overal Passing ']]

District_Summary = District_Summary.round(2)
```
